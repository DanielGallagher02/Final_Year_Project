```python
# Dataset exploration of wheat Dataset'Wheat1_'
import matplotlib.pyplot as pyplot
from PIL import Image, ImageFilter
import os

# Assuming your wheat images are stored in the folder named 'Wheat1_' inside 'data/raw'
wheat_images_dir = '../data/raw/Wheat1_'
wheat_image_files = [f for f in os.listdir(wheat_images_dir) if f.endswith('.jpeg')]

# Display the first image in the list
first_image_path = os.path.join(wheat_images_dir, wheat_image_files[0])
image = Image.open(first_image_path)
pyplot.imshow(image)
pyplot.title(wheat_image_files[0])
pyplot.axis('off') # Hide the axis
pyplot.show()

# An Example of transforming the image: rotating it by 90 degrees
rotated_image = image.rotate(90)
pyplot.imshow(rotated_image)
pyplot.title('Rotated Image')
pyplot.axis('off')
pyplot.show()

# An Example of filtering the image: applying a blur filter
blurred_image = image.filter(ImageFilter.BLUR)
pyplot.imshow(blurred_image)
pyplot.title('Blurred Image')
pyplot.axis('off')
pyplot.show()

# An Example of image analysis: displaying the histogram of the image
image_histogram = image.histogram()
pyplot.plot(image_histogram)
pyplot.title('Image Histogram')
pyplot.show()




```
